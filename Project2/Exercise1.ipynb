```python
# Necessary imports
import numpy as np
from matplotlib import pyplot as plt
```

```python
# necessary lists
substrate_conc = np.array([0.2, 0.4, 0.8, 1.0, 2.0, 4.0])
plot_s = 1/substrate_conc

# inhibitor kinetics
no_inhib = np.array([5.0, 7.5, 10.0, 10.7, 12.5, 13.6])
inhib_A = np.array([3.0, 5.0, 7.5, 8.3, 10.7, 12.5])
inhib_B = np.array([2.0, 3.0, 4.0, 4.3, 5.0, 5.5])

plot_no_inhib = 1/no_inhib
plot_inhib_a = 1/inhib_A
plot_inhib_b = 1/inhib_B
```

Plot for enzyme kinetics for inhibitor A

```python
plt.figure()
plt.grid(True)
plt.plot(plot_s, plot_inhib_a,'.-', plot_s, plot_no_inhib, '.-')
plt.title('Lineweaver-Burk Plot for inhibitor A')
plt.legend(['inhibitor A', 'no inhibitor'])
plt.xlabel('1/[S] (1/mM)')
plt.ylabel('1/v (min/mM)')
plt.show()

# Linear fitting and calculation of kinetic parameters - no inhibitor
slope_no, intercept_no = np.polyfit(plot_s, plot_no_inhib, 1)
Vmax_no = 1/intercept_no
Km_no = slope_no*Vmax_no
print("Vmax for no inhibitor:", Vmax_no, "mM/min")
print("Km for no inhibitor", Km_no, "mM")

# Linear fitting and calculation of kinetic parameters - Inhibitor A
slope_a, intercept_a = np.polyfit(plot_s, plot_inhib_a, 1)
Vmax_a = 1/intercept_a
Km_a = slope_a*Vmax_a
print("Vmax for inhibitor A:", Vmax_a, "mM/min")
print("Km for inhibitor A", Km_a, "mM")
```

Plot for inhibitor B

```python
plt.figure()
plt.grid(True)
plt.plot(plot_s, plot_inhib_b,'.-', plot_s, plot_no_inhib, '.-')
plt.title('Lineweaver-Burk Plot for inhibitor B')
plt.legend(['inhibitor B', 'no inhibitor'])
plt.xlabel('1/[S] (1/mM)')
plt.ylabel('1/v (min/mM)')
plt.show()

slope_b, intercept_b = np.polyfit(plot_s, plot_inhib_b, 1)
Vmax_b = 1/intercept_b
Km_b = slope_b*Vmax_b
print("Vmax for inhibitor B:", Vmax_b, "mM/min")
print("Km for inhibitor B", Km_b, "mM")
```

Question (c)

For inhibitor A, we observe that Vmax remains unchanged while Km doubles compared to the kinetics with no inhibitor. This corresponds to competitive inhibition meaning that the inhibitor ccompetes with the substrate for the active site. More substarte is hence required to reach maximum velocity which results in an increased Km. Mathematically, taking the kinetics equation of the competitive inhibition: 

$$v = \frac{V_{max}*S}{K_m*(1+\frac{I}{K_I})+S}$$

In the limit of the substrate concentration approaching infinity, we get:

$$v = V_{max}$$

Hence inhibitor A corresponds well to the competitive inhibition. 

As for inhibitor B, we observe that this time, Vmax gets reduced while Km stays the same when compared to no inhibition case. This corresponds to an uncompetitive inhibition. According to the uncompetitive inhibition equation: 

$$\frac{V_{max}*S}{K_{m}+S(1+\frac{I}{K_{I}})}

Once again, when S approaches infinity, the equation becomes: 

$$\frac{V_{max}}{1+\frac{I}{K_I}}$$

Hence Vmax is reduced. It is important to note that in case of a non-competitive inhibition, Vmax is also reduced. However, we distinguish inhibitor B from a non-competitive inhibitor using the fact that, in a non-competitive inhibition, Km is also reduced which is not our case.

```python
# DATA 
#conc
s = np.array([0.1, 0.3, 0.7, 2.0, 4.0, 6.0])
inv_S = 1/s # Lineweaver-Burk transformation

# rate
v1 = np.array([1.09, 2.89, 5.24, 8.85, 10.89, 12.11])
v2 = np.array([0.74, 1.96,3.89, 7.40, 9.82, 11.36])
v3 = np.array([0.52, 1.47, 3.01, 6.31, 8.76, 10.02])
v4 = np.array([0.42, 1.17, 2.51, 5.43, 8.07, 9.37])

inv_v1 = 1/v1
inv_v2 = 1/v2
inv_v3 = 1/v3
inv_v4 = 1/v4
```

```python
# Plot
plt.figure()
plt.title('Lineweaver-Burk of inhibitor kinetics under different concentrations')
plt.grid(True)
plt.plot(inv_S, inv_v1, '.-',inv_S, inv_v2, '.-',inv_S, inv_v3, '.-',inv_S, inv_v4, '.-')
plt.legend(['[I] = 1mM','[I] = 2mM','[I] = 3mM','[I] = 4mM'])
plt.xlabel('1/[S] (1/mM)')
plt.ylabel('1/v (min/mM)')
plt.show()
```

By visually inspecting the plot, we can make the hypothesis that the inhibitor is competitive on the basis that the y-intercept seems to remain approximately constant which means that Vmax will remain constant accross all concentrations. The slope, however, increases with increasing concentration indicating a change in Km. 

To estimate Ki we use the following equation: 

$$slope = \frac{K_m}{V_{max}}(1+\frac{[I]}{K_I})

```python
# Estimation of Ki

# Concentrations of the inhibitor 
i = np.array([1,2,3,4])
fits = [inv_v1, inv_v2, inv_v3, inv_v4]
slopes = []

for j in fits:
    slope, intercept = np.polyfit(inv_S, j, 1)
    slopes.append(slope)

m, c = np.polyfit(i, slopes, 1)
Ki = c/m
print("Ki: ", Ki, "mM")
```
